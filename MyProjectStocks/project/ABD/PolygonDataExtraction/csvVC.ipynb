```python
import pandas as pd
import requests
import time
```

```python
api_key = 'Jpdzf05nPJhqVB5qpNZx2VbmhwVF1spP'
```

```python
df = pd.read_csv('/Users/tbai/Documents/Vsanaliz/analiz/project/BistTum/sp500_symbols.csv')  
bist_tum_symbols = df.iloc[:, 0].tolist()
```

```python
for symbol in bist_tum_symbols:
    url = f'https://api.polygon.io/v1/meta/symbols/{symbol}/company?apiKey={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        print(f"{symbol} geçerli bir sembol.")
    else:
        print(f"{symbol} geçerli değil.")
```

```python
bist_data = {}
failed_symbols = {}
nan_data = {} 

start_date = "2023-08-20"
end_date = "2024-08-20"

for symbol in bist_tum_symbols:
    try:
        print(f"{symbol} için veri çekiliyor...")

        url = f'https://api.polygon.io/v2/aggs/ticker/{symbol}/range/1/day/{start_date}/{end_date}?adjusted=true&sort=asc&limit=50000&apiKey={api_key}'
        response = requests.get(url)
        data = response.json()

        if response.status_code == 200 and 'results' in data:
            df = pd.DataFrame(data['results'])
            df['date'] = pd.to_datetime(df['t'], unit='ms') 
            
            if not df.empty and not df.isnull().values.any():
                bist_data[symbol] = df
            else:
                print(f"Veri NaN içeriyor veya eksik veri var: {symbol}")
                nan_data[symbol] = df
                failed_symbols[symbol] = "Veri NaN içeriyor veya eksik veri var"
        else:
            print(f"Veri çekilemedi: {symbol}")
            failed_symbols[symbol] = "Veri çekilemedi"
 

    except Exception as e:
        print(f"Hata: {symbol} - {e}")
        failed_symbols[symbol] = str(e)

if bist_data:
    bist_combined = pd.concat(bist_data, axis=1, keys=bist_data.keys())
    
    bist_combined_cleaned = bist_combined.dropna(axis=1)
    
    print(bist_combined_cleaned.head())
    bist_combined_cleaned.to_csv('/Users/tbai/Documents/Vsanaliz/analiz/BistTumVeriler_Polygon.csv')
    print("Veriler başarıyla kayıt edildi.")

if nan_data:
    nan_combined = pd.concat(nan_data, axis=1, keys=nan_data.keys())
    nan_combined.to_csv('/Users/tbai/Documents/Vsanaliz/analiz/BistTumVeriler_NaN_Polygon.csv')
    print("NaN içeren veriler ayrı olarak kaydedildi.")

if failed_symbols:
    print("İndirilemeyen veya eksik veri içeren hisse kodları ve hatalar:")
    for symbol, error in failed_symbols.items():
        print(f"{symbol}: {error}")
    
    failed_df = pd.DataFrame(list(failed_symbols.items()), columns=['Symbol', 'Error'])
    failed_df.to_csv('/Users/tbai/Documents/Vsanaliz/analiz/FailedSymbols_Polygon.csv', index=False)
    print("Hatalı hisse kodları kayıt edildi.")
    
print("işlem bitti")
```
